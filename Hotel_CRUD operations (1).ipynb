```python
import pandas as pd
from pymongo import MongoClient
import json

clients = MongoClient("mongodb://localhost:27017/")

#function to convert csv to json and upload to mongodb server
def mongo_upload(csv_path, db_name, coll_name):
    dbs = clients[db_name]
    collection = dbs[coll_name]
    dataset = pd.read_csv(csv_path)
    json_data = json.loads(dataset.to_json(orient='records'))
    collection.drop() #prevents writing the same data multiple times to the collection
    collection.insert_many(json_data)
    return "Success"
#     return coll.count_documents({})
```
